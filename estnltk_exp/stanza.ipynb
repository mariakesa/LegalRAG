```python
import stanza

# Initialize the Stanza pipeline
stanza.download('en') # Download English model if you haven't
nlp = stanza.Pipeline('en') # Initialize pipeline

# Define your legal text
legal_text = "John Doe is suing Jane Smith for breach of contract."

# Process the document with Stanza
doc = nlp(legal_text)

# Example: Find who is suing who based on verb and dependency relations
for sentence in doc.sentences:
    for word in sentence.words:
        if word.text == 'suing':  # Look for specific verbs like 'suing'
            subject = [w.text for w in sentence.words if w.head == word.id and w.deprel == 'nsubj']
            object_ = [w.text for w in sentence.words if w.head == word.id and w.deprel == 'obj']
            if subject and object_:
                print(f"{subject[0]} is suing {object_[0]}")
```

```python
import spacy
from spacy.pipeline import EntityRuler

# Load a spaCy model (English in this case)
nlp = spacy.load('en_core_web_sm')

# Create an EntityRuler and add it to the pipeline by its name
ruler = nlp.add_pipe("entity_ruler")

# Define your custom legal lexicon
legal_lexicon = [
    {"label": "PLAINTIFF", "pattern": "plaintiff"},
    {"label": "DEFENDANT", "pattern": "defendant"},
    {"label": "COURT", "pattern": "court"},
    # Add more legal terms...
]

# Add patterns to the EntityRuler
ruler.add_patterns(legal_lexicon)

# Example legal text
legal_text = "The plaintiff, John Doe, sued the defendant, Jane Smith, in court over a contract dispute."

# Process the text with the pipeline
doc = nlp(legal_text)

# Print detected entities
for ent in doc.ents:
    print(ent.text, ent.label_)
```
