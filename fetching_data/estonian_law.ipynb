```python
import requests
from bs4 import BeautifulSoup

# URL of the webpage
kars = "https://www.riigiteataja.ee/akt/184411"

# Send a request to fetch the content of the webpage
r = requests.get(kars)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(r.content, 'html.parser')

# Extract all text from the webpage
text = soup.get_text()

# Optionally, clean up the text by stripping extra spaces
cleaned_text = " ".join(text.split())

# Print the extracted text
print(cleaned_text)
```

```python
import requests
from bs4 import BeautifulSoup
import spacy
from IPython.display import display, Markdown

# Load the multilingual spacy model that supports Estonian
nlp = spacy.load("xx_sent_ud_sm")

# URL of the Estonian webpage
kars = "https://www.riigiteataja.ee/akt/184411"

# Send a request to fetch the content of the webpage
r = requests.get(kars)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(r.content, 'html.parser')

# Extract all text from the webpage
text = soup.get_text()

# Optionally, clean up the text by stripping extra spaces
cleaned_text = " ".join(text.split())

# Use spacy to process the Estonian text
doc = nlp(cleaned_text)

# Extract sentences
sentences = [sent.text for sent in doc.sents]

# Display the sentences in the Jupyter notebook
for sentence in sentences:
    display(Markdown(sentence))
```
